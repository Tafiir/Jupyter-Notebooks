```python
m = arange(3,13)
l3 = 819.6
dl3 = 3.7
E3p = -13.605/9 - 1.2398e+3/l3
dE3p = 1.2398e+3/(l3**2) * dl3
valerr("E_{{3p}}", E3p, dE3p, 3,"eV")
l_t = 1.2398e+3/(-13.605/(m**2) - E3p)
dl_t = l_t * dE3p/(-13.605/(m**2) - E3p)


l_f = array([819.6, None, 497.6, 466.4, 449.2, None, 433.0, 429.4, 426.1, 419.3], dtype=float64).round(1)
dl_f = []

from scipy.interpolate import UnivariateSpline
mask = na_li2[0].map(lambda x: x in arange(l_f[0]-5, l_f[0]+5,0.01, dtype=float).round(2))
spline = UnivariateSpline(na_li2[mask][0], na_li2[mask][1]-(na_li2[mask][1].max()+na_li2[mask][1].min())/2, s=0)
dl_f.append(spline.roots()[1]-spline.roots()[0])

for i in range(len(l_f)):
    if not isnan(l_f[i]) and not i==0:
        mask = na_li1[0].map(lambda x: x in arange(l_f[i]-5, l_f[i]+5,0.01, dtype=float).round(2))
        spline = UnivariateSpline(na_li1[mask][0], na_li1[mask][1]-(na_li1[mask][1].max()+na_li1[mask][1].min())/2, s=0)
        root = spline.roots()
        dl_f.append(root[1]-root[0])
    elif i != 0:
        dl_f.append(None)

na_lines = pd.DataFrame.from_items([("m",m),
                                    ("$\lambda_{{theoretisch}} \ [nm]$",l_t),
                                    ("$\Delta \lambda_{{theoretisch}}$",dl_t),
                                    ("$\lambda_{{gefunden}} \ [nm]$",l_f),
                                    ("$\Delta \lambda_{{gefunden}}$",dl_f)
                                   ]).set_index("m")
    
display(na_lines.round(1))

lm = lambda m, Er, Ep, D: 1.2398e+3/(Er/((m-D)**2) - Ep)
lm_ = lambda m, Ep, D: 1.2398e+3/(-13.605/((m-D)**2) - Ep)

from scipy.optimize import curve_fit
popt, pcov = curve_fit(lm, na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
                       sigma=na_lines.dropna().iloc[:,3], p0=[-13.6, -3.02, 0.08])
display(popt, pcov)

figure()
errorbar(na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
         yerr=na_lines.dropna().iloc[:,3].values, fmt='bs')
plot(arange(3,12,0.01), lm(arange(3,12,0.01), *popt))
xlim((2.9,12.1))
xticks(arange(3,13))
xlabel(r"m")
ylabel(r"$\lambda_m \ [nm]$")
grid()
show()

valerr("E_{{Ry}}", popt[0], sqrt(pcov[0][0]), 4, "eV")
valerr("E_{{3p}}", popt[1], sqrt(pcov[1][1]), 4, "eV")
valerr("\Delta_d", popt[2], sqrt(pcov[2][2]), 4)

popt, pcov = curve_fit(lm_, na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
                       sigma=na_lines.dropna().iloc[:,3], p0=[-3.02, 0.8])

figure()
errorbar(na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
         yerr=na_lines.dropna().iloc[:,3].values, fmt='bs')
plot(arange(3,12,0.01), lm_(arange(3,12,0.01), *popt))
xlim((2.9,12.1))
xticks(arange(3,13))
xlabel(r"m")
ylabel(r"$\lambda_m \ [nm]$")
grid()
show()
valerr("E_{{3p}}", popt[0], sqrt(pcov[0][0]), 4, "eV")
valerr("\Delta_d", popt[1], sqrt(pcov[1][1]), 4)
```

```python
m = arange(4, 10)
l_D = 588.8
mask = na_hi[0].map(lambda x: x in arange(580, 605,0.01, dtype=float).round(2))
spline = UnivariateSpline(na_hi[mask][0], na_hi[mask][1]-(na_hi[mask][1].max()+na_hi[mask][1].min())/2, s=0)
root = spline.roots()
dl_D = root[1]-root[0]
valerr("\lambda_D", l_D, dl_D, 2, "nm")

E3s = E3p - 1.2398e+3/l_D
dE3s = sqrt(dE3p**2 + (1.2398e+3/(l_D**2) * dl_D)**2)
valerr("E_{{3s}}", E3s, dE3s, 4, "eV")
Ds = -sqrt(-13.605/E3s) + 3
dDs = -sqrt(-13.605/E3s)/(E3s*2) * dE3s
valerr("\Delta_s", Ds, dDs, 4)
l_t = 1.2398e+3/(-13.605/((m-Ds)**2) - E3p)
dl_t = sqrt((2*1.2398e+3*(-13.605)*(m-Ds)/(((-13.605) - ((m-Ds)**2)*E3p)**2) * dDs)**2 + (1.2398e+3/((-13.605/((m-Ds)**2) - E3p)**2) * dE3p)**2)

l_f = array([None, 615.8, 514.7, 474.5, 454.9,438.5], dtype=float64).round(1)
dl_f = []

dl_f.append(None)
mask = na_li2[0].map(lambda x: x in arange(l_f[1]-5, l_f[1]+5,0.01, dtype=float).round(2))
spline = UnivariateSpline(na_li2[mask][0], na_li2[mask][1]-(na_li2[mask][1].max()+na_li2[mask][1].min())/2, s=0)
root = spline.roots()
dl_f.append(root[1]-root[0])

for i in range(len(l_f)):
    if not isnan(l_f[i]) and i>1:
        mask = na_li1[0].map(lambda x: x in arange(l_f[i]-5, l_f[i]+5,0.01, dtype=float).round(2))
        spline = UnivariateSpline(na_li1[mask][0], na_li1[mask][1]-(na_li1[mask][1].max()+na_li1[mask][1].min())/2, s=0)
        root = spline.roots()
        if len(root)>1:
            dl_f.append(root[1]-root[0])
        else:
            dl_f.append((l_f[i]-root[0])/2)

na_lines = pd.DataFrame.from_items([("m",m),
                                    ("$\lambda_{{theoretisch}} \ [nm]$",l_t),
                                    ("$\Delta \lambda_{{theoretisch}}$",dl_t),
                                    ("$\lambda_{{gefunden}} \ [nm]$",l_f),
                                    ("$\Delta \lambda_{{gefunden}}$",array(dl_f, dtype=float64))
                                   ]).set_index("m")

display(na_lines.round(1))

popt, pcov = curve_fit(lm, na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
                       sigma=na_lines.dropna().iloc[:,3].values,p0=[-13.6, -3.02, 1.3])

figure()
errorbar(na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
         yerr=na_lines.dropna().iloc[:,3].values, fmt='bs')
plot(arange(4,10,0.01), lm(arange(4,10,0.01), *popt))
xlim((3.9,9.1))
xticks(arange(4,10))
xlabel(r"m")
ylabel(r"$\lambda_m \ [nm]$")
grid()
show()

valerr("E_{{Ry}}", popt[0], sqrt(pcov[0][0]), 4, "eV")
valerr("E_{{3p}}", popt[1], sqrt(pcov[1][1]), 4, "eV")
valerr("\Delta_s", popt[2], sqrt(pcov[2][2]), 4)

popt, pcov = curve_fit(lm_, na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
                       sigma=na_lines.dropna().iloc[:,3].values, p0=[-3.02, 0.8])

figure()
errorbar(na_lines.dropna().index.values, na_lines.dropna().iloc[:,2],
         yerr=na_lines.dropna().iloc[:,3].values, fmt='bs')
plot(arange(4,10,0.01), lm_(arange(4,10,0.01), *popt))
xlim((3.9,9.1))
xticks(arange(4,10))
xlabel(r"m")
ylabel(r"$\lambda_m \ [nm]$")
grid()
show()

valerr("E_{{3p}}", popt[0], sqrt(pcov[0][0]), 4, "eV")
valerr("\Delta_s", popt[1], sqrt(pcov[1][1]), 4)
```
