```python
f_ = array([330,323,327])/60
df = float64(1/60)

f = f_.mean()

valerr("f", f, df, 3, "s^{{-1}}")

I = 2.53 * 5
dI = 0.17 * 5

U = 11.5
dU = 1.1

Q_el = U*I/f
dQ_el = U*I/f*sqrt((dU/U)**2 + (dI/I)**2 + (df/f)**2)

valerr("P_{{el}}", U*I, sqrt((dU*I)**2 + (dI*U)**2), 0, "W")
valerr("Q_{{el}}", U*I/f, U*I/f*sqrt((dU/U)**2 + (dI/I)**2 + (df/f)**2), 1, "J")

T_ab = 22.5
dT = 0.1
T_zu = 17.2

cw = 4180
p = 998

V = 257e-6/60
dV = 2e-6/60
valerr("\dot{{V}}", V, dV)

P_ab = float64(cw * p * (T_ab - T_zu) * V)
dP_ab = P_ab * sqrt((dV/V)**2 + 2*(dT/(T_ab-T_zu))**2)

Q_ab = P_ab/f
dQ_ab = Q_ab * sqrt((dP_ab/P_ab)**2 + (df/f)**2)

valerr("P_{{ab}}", P_ab, dP_ab, 1, "W")
valerr("Q_{{ab}}", Q_ab, dQ_ab, 2, "J")

A_ = array([19857, 20328, 20014]) * 10**(-4)
Q_pv = A_.mean()
dQ_pv = A_.std()
P_pv = Q_pv*f
dP_pv = sqrt((df*Q_pv)**2 + (f*dQ_pv)**2)

valerr("P_{{pV}}", P_pv, dP_pv, 2, "W")
valerr("Q_{{pV}}", Q_pv, dQ_pv, 3, "J")

n_th = Q_pv/Q_el
dn_th = n_th * sqrt((dQ_pv/Q_pv)**2 + (dQ_el/Q_el)**2)

valerr("\eta_{{th}}", n_th, dn_th, 4)

valerr("Q_V",(Q_el - Q_ab - Q_pv), sqrt(dQ_el**2 + dQ_ab**2 + dQ_pv**2), 1, "J")
```

```python
l = 0.25

F = array([0.8, 0.6, 0.4, 0.2])
dF = array([0.1, 0.05, 0.05, 0.1])

D = l * F
dD = l * dF

W_D = D * 2*pi
dW_D = dD * 2*pi

A = array([[21884, 23322, 24976],
           [24471, 24636, 24119],
           [23499, 23575, 23512],
           [22450, 21824, 21965]]) * 10**(-4)
W_pv = A.mean(axis=1)
dW_pv = A.std(axis=1)

U = array([11.47, 11.48, 11.44, 11.48])
dU = 1.1

I = array([2.53, 2.54, 2.55, 2.54]) * 5
dI = 0.17 * 5

f_ = array([[205, 216, 211],
           [249, 241, 243],
           [282, 280, 281],
           [315, 314, 316]])/60
f = f_.mean(axis=1)
df = float64(1/60) * sqrt(3)

Q_el = U * I / f
dQ_el = Q_el * sqrt((dU/U)**2 + (dI/I)**2 + (df/f)**2)

out = pd.DataFrame.from_items([("$W_D \ [J]$", W_D),
                               ("$\Delta W_D$", dW_D),
                               ("$W_{{pV}} \ [J]$", W_pv),
                               ("$\Delta W_{{pV}}$", dW_pv),
                               ("$Q_{{el}}$", Q_el),
                               ("$\Delta Q_{{el}}$", dQ_el),
                               ("$Q_{{el}}$", f),
                               ("$\Delta Q_{{el}}$", df)
                              ])
display(out.round(4))
```
