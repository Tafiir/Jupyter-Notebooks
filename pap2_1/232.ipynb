```python
from sqlalchemy import create_engine

db = create_engine("sqlite:///232.db")

data2 = pd.read_sql('wavelength', db)

ds1 = 9
dm = 2
display((data2.iloc[:,1]-data2.iloc[:,0]).std())
display(sqrt((data2.iloc[:,1]-data2.iloc[:,0]).std()**2 + ds1**2))

ds = sqrt((data2.iloc[:,1]-data2.iloc[:,0]).std()**2 + ds1**2)

l_ = 2*(data2.iloc[:,1]-data2.iloc[:,0])/data2.iloc[:,2]
dl = l_ * sqrt((ds/(data2.iloc[:,1]-data2.iloc[:,0]))**2 + (dm/data2.iloc[:,2])**2)

display(pd.DataFrame.from_items([("$\lambda \ [\mu m]$",l_), ("$\Delta \lambda$",dl)]))
display(Latex(r"$\lambda = {0:.1f}\pm {1:.1f} nm$".format(l_.mean() * 10**3, sqrt(sum(dl**2))/len(dl) * 10**3)))
```

```python
data3 = pd.read_sql('pressure', db)

dp = sqrt((0.006*data3.iloc[:,1:])**2 + 5)
dpm = dp.mean(axis=1)

pm = data3.iloc[:,1:].mean(axis=1)
display(pd.DataFrame.from_items([("$\overline{{p}} \ [Torr]$",pm), ("$\Delta \overline{{p}}$",dpm)]))

from scipy.odr import RealData, Model, ODR
dat = RealData(x=pm, y=data3.iloc[:,0], sx=dpm)
f = lambda p, x: p[0]*x +p[1]
modf = Model(f)
odr = ODR(dat, modf, [-1,46])
out = odr.run()
out.pprint()


figure()
errorbar(pm, data3.iloc[:,0], xerr=dpm, fmt='bs')
plot(arange(100,800,0.001), f(out.beta, arange(100,800,0.001)))
xlabel(r"$\overline{{p}} \ [Torr]$")
ylabel(r"$m$")
show()
```

```python
from detect_peaks import detect_peaks

pind = detect_peaks(data4.iloc[7250:9000,1], mph=0.0001, mpd=20) + 7250
```

```python
from scipy.optimize import curve_fit

gauss = lambda t, a, u, s, y0: a/(s*sqrt(2*pi)) * exp(-(t-u)**2/(2*s**2)) +  y0

popt, pcov = curve_fit(gauss, data4.iloc[pind,0], data4.iloc[pind,1], p0=[0.008, 0.015, 0.02, 0.001])
display(Latex(r"$a = {0:.4f} \\ \mu = {1:.4f} \\ \sigma = {2:.4f} \\y_0 = {3:.4f}$".format(*popt)))
display(Latex(r"$\Delta a = {0:.4f} \\ \Delta \mu = {1:.4f} \\ \Delta \sigma = {2:.4f}$".format(sqrt(pcov[0][0]),
                                                                                                sqrt(pcov[1][1]),
                                                                                                sqrt(pcov[2][2]),
                                                                                                sqrt(pcov[2][2]))))

figure()
plot(data4.iloc[:,0], data4.iloc[:,1], label=r"Spannung")
# scatter(data4.iloc[pind,0], data4.iloc[pind,1], c='r',marker='+')
plot(data4.iloc[:,0], gauss(data4.iloc[:,0], *popt), 'r-', label=r"Gauß Fit")
xlim((-0.06, 0.07))
xlabel(r"Zeit $[s]$")
ylabel(r"Spannung $[V]$")
minorticks_on()
grid()
show()
```
