```python
data = pd.read_csv('232.csv')
data.drop('Unnamed: 2',axis=1, inplace=True)
data.columns = data.columns + ' [' + data.iloc[0,:] + ']'
data = data.convert_objects(convert_numeric='force')
data = data.drop([0]).reset_index().drop('index',axis=1)
```

```python
fig = figure(figsize=(6,8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(data.iloc[:,0], data.iloc[:,1])
ax1.set_xlim((-0.06, 0.07))
ax1.grid()
ax1.minorticks_on()

ax2.plot(range(len(data.iloc[:,0])), data.iloc[:,1])
ax2.set_xlim((7250, 9000))
ax2.minorticks_on()
ax2.grid()

show()
```

```python
from detect_peaks import detect_peaks

pind = detect_peaks(data.iloc[7250:9000,1], mph=0, mpd=20) + 7250

figure()
plot(data.iloc[:,0], data.iloc[:,1])
scatter(data.iloc[pind,0], data.iloc[pind,1], c='r',marker='+')
xlim((-0.06, 0.07))
minorticks_on()
grid()
show()
```

```python
from scipy.optimize import curve_fit

gauss = lambda t, a, u, s, y: a/(s*sqrt(2*pi)) * exp(-(t-u)**2/(2*s**2)) + y

popt, pcov = curve_fit(gauss, data.iloc[pind,0], data.iloc[pind,1], p0=[0.008, 0.015, 0.02, 0.0001])
display(Latex(r"$a = {0:.4f} \\ \mu = {1:.4f} \\ \sigma = {2:.4f} \\ y_0 = {3:.4f}$".format(*popt)))

figure()
plot(data.iloc[:,0], data.iloc[:,1])
# scatter(data.iloc[pind,0], data.iloc[pind,1], c='r',marker='+')
plot(data.iloc[:,0], gauss(data.iloc[:,0], *popt), 'r-')
xlim((-0.06, 0.07))
grid()
show()
```
